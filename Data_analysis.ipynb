Data scrubbing

```python
churn_dfnew=churn_df.copy()
churn_dfnew.info()
```

```python
# converting total charge from object to numeric
churn_dfnew.TotalCharges=pd.to_numeric(churn_dfnew.TotalCharges,errors='coerce')
```

```python
churn_dfnew.isnull().sum()
```

```python
churn_dfnew.loc[churn_dfnew.TotalCharges.isnull()== True]
```

```python
# so missing values are too low, so dropping them
churn_dfnew.dropna(how='any',inplace=True)
```

```python
churn_dfnew.shape
```

```python
# columns not needed for further processing to be dropped
churn_dfnew.drop(columns=['customerID','tenure'],axis=1,inplace=True)
```

```python
churn_dfnew.shape
```

Univariate analysis

```python
for i, predictor in enumerate(churn_dfnew.drop(columns=['Churn','MonthlyCharges','TotalCharges'])):
  plt.figure(i,figsize=(12,8))
  sns.countplot(data=churn_dfnew,x=predictor,hue='Churn')
```

```python
# converting the output(churn) as 0,1
churn_dfnew['Churn']=np.where(churn_dfnew.Churn=='Yes',1,0)
```

```python
churn_dfnew.head()
```

```python
# convering categorical values to dummy
churn_dfdummy=pd.get_dummies(churn_dfnew)
churn_dfdummy.shape
```

```python
plt.figure(figsize=(10,8))
monCharge=sns.kdeplot(churn_dfdummy.MonthlyCharges[(churn_dfdummy["Churn"]==0)],color='green',shade=True)
monCharge=sns.kdeplot(churn_dfdummy.MonthlyCharges[(churn_dfdummy["Churn"]==1)],color='red',shade=True)
monCharge.set_title('Churning with respect to monthly charges')
monCharge.set_xlabel('monthly charges')
monCharge.set_ylabel('density')
monCharge.legend(["Not churn","Churn"],loc='upper left')
```

The above plot tells that high monthly charges cause high churning.

```python
plt.figure(figsize=(20,12))
churn_dfdummy.corr()['Churn'].sort_values().plot(kind='bar')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(churn_dfdummy.corr(),cmap='Blues')
```

Above correlation suggests that high monthly charges,no technical support, no online security, no online backup results in high churning
